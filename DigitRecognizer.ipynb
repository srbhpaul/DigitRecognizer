## Understanding the kaggle training set

### Number of digits

We will first understand the training data set provided from kaggle. It contains 10 (0,1,2,...9) digits, with a total of 42000 examples. We will actually use a subset of these large set as our training set. Our first job is to find out the number(percentage) of examples corresponding to each digit in this training set provided from kaggle. This will enable us to check if there is any large bias in the training set, and might as well give us ideas of properly constructing a subset of the training set. 

```python
# no. of classifiers or digits (0,1,2,...9)
k = 10
# read the original kaggle traning set as a DataFrame object
train_original = pd.read_csv('train.csv')

# the feature and label matrices, X and Y, respectively
# to create the featire matrix X, we add a first column of ones
train_original_mat = train_original.as_matrix()
Xtrain_original = train_original_mat[:,1:]
Xtrain_original = np.insert(Xtrain_original,0,1,axis=1)
Ytrain_original = train_original_mat[:,0]

# find out the number of examples for each digit
digits = np.arange(k)
digit_count = np.zeros(k,int)
for i in xrange(train_original.shape[0]):
    for j in xrange(k):
        if train_original.label[i] == digits[j]:
            digit_count[j]+=1
digit_percentage = ( digit_count/float(np.sum(digit_count)) )*100.0
```

```python
# plot digit_count and digit_percentage

plt.figure(figsize=(10, 3.5))

# plot 1
plt.subplot(1,2,1)
plt.xlabel('digits',fontsize=16)
plt.ylabel('count',fontsize=16)
plt.xlim(-0.5,9.5)
plt.ylim(3650,4750)
plt.plot(digits,digit_count,marker='o')

# plot 2
plt.subplot(1,2,2)
plt.xlabel('digits',fontsize=16)
plt.ylabel('% count',fontsize=16)
plt.xlim(-0.5,9.5)
plt.ylim(8.5,11.5)
plt.plot(digits,digit_percentage,marker='o')

plt.tight_layout(w_pad=8.0)
```

We thus see that the original training set from kaggle has relatively fewer training examples for digit 5, while a much larger set for digit 1.

### Visualizing the digits

Each digit is represented as a grey-scale image of 28x28 pixels. Each row of the training set contains 28x28=784 pixel values. Below, we define a function that plots random images from the training set in a grid of 8x8 images.

```python
# plot digits for an arbitrary input matrix X
# the input has to be a numpy array
# the function outputs a figure of 8x8 random digits 
# from the input array

def plot_digits(X):
    # image size = imglen x imglen 
    imglen = int(sqrt(X.shape[1]))
              
    # enter the number of images in each row, 
    # i.e., a figure of grid size imageno x imageno
    imageno = 8           
    
    # plotting the digits!
    plt.figure(figsize=(6,6))
    for i in xrange(imageno):
        for j in xrange(imageno):
            j += imageno*i 
            # example image, i.e., select a random row of X
            eximg = np.random.random_integers(0,X.shape[0]) 
            imgdata = X[eximg,:].reshape((imglen,imglen))
            plt.subplot(8,8,j)
            plt.imshow(imgdata,cmap = plt.cm.Greys_r)
            plt.axis('off')
        plt.tight_layout(h_pad=-3.5)
        plt.tight_layout(w_pad=-1.5)  
```

## Logistic regression using an unbiased training set

We use logistic regression to predict the digits in the test set. 

### Unbiased training set 

We create a smaller training set for faster numerical computation, and also to check how the construction of a training set affects the prediction accuracy. Our first attempt is to construct a training set containing equal number of training examples of every digit. Let the chosen number of training examples per digit be $n$. We construct this training set by choosing the first $n$ occurances of every digit from the kaggle training set. The total number of training examples used is thus $N = nk$, where $k=10$ is the number of labels, or in this case, digits. We properly rescale the pixel values of the training set for faster numerical convergence. 

### Test set

We treat the entire original kaggle traning set with $M = 42000$ examples as our test set. This is justified since we will be interested in $N/M\ll 1$, i.e., the truncated training set is much smaller than the original kaggle training set ensuring that the test set has negligible overlap with the training set. Choosing a labeled test set allows us to check the prediction accuracy of our model. We again appropriately rescale the pixel values in the test set, as done previously in creating the truncated training set.

```python
# rescaling the pixel values to create the X matrix
# for the test data set, which is the entire kaggle training set

num = train_original.shape[0]
testdata = train_original
Xtest = testdata.ix[:,1:]
Xtest_mean = Xtest.mean(axis=1)
Xtest_std = Xtest.std(axis=1)

for i in xrange(num):
    Xtest.ix[i,:] = (Xtest.ix[i,:]-Xtest_mean.ix[i])/Xtest_std.ix[i]
    i+=1
Xtest = Xtest.as_matrix()

# adding the column of ones to create the feature matrix
Xtest = np.insert(Xtest,0,1,axis=1)
# label matrix
Ytest = testdata.ix[:,0].as_matrix()
```

### Predictions from the model

Obtaining the percentage accuracy of predictions from the logistic regression, along with the predicted digits and corresponding probabilities. We write this as a general class, that takes predicted logistic regression $\theta$ as parameters, feature matrix $X$ and classifier matrix $Y$ as instances. It outputs the various quantities as members of the class 'predict'.

```python
# defining the various basic quantities 
# that we would like to predict
# and measure the accuracy of our predictions 
# using logistic regression
# inputs X and Y are arbitrary 
# feature and label matrix
# both must be numpy arrays
# For our purposes, X = Xtest, Y = Ytest
# input theta is the trained parameter set 
# from logistic regression
# theta is also a numpy array
# this is the most generl such class
class Predict:
    def __init__(self,X,Y):
        self.X, self.Y = X,Y
        
    # Probabilistic predictions for the 
    # feature matrix X, a numpy array. 
    # the corresponding classifiers or digits 
    # are in the matrix Y, a numpy array
    def probs(self,theta):
        return lgs.g( np.dot( self.X,np.transpose(theta) ) )
    
    def pred_digits(self,theta):
        # The predicted digit is the 
        # classifier with maximum probability.
        # returns an array of size Xtest.shape[0], 
        # containing predicted digits
        return  np.argmax( self.probs(theta), axis=1 )
     
    # counting the number of examples for 
    # every digit in the input label set Y
    def digit_count(self):
        # counting the number of occurances 
        # of individual digits in the Y matrix
        k = 10 # number or classifiers or digits
        acc = np.zeros(k,float)
        count = np.zeros(k,int)
        digits = np.arange(k)
        for i in xrange(self.Y.shape[0]):
            for j in xrange(k):
                if self.Y[i] == digits[j]:
                    count[j]+=1
        return count
```

Predictions for the unbiased training set. We inherit from the base class Predict, and add relevant member functions.

```python
# class for the unbiased training case
# we define how to calculate accuracy in this case

class Predict_Unbiased(Predict):
                   
    def accuracy(self,theta):        
        # predicted digits
        predict_digits = self.pred_digits(theta)
        
        # overall prediction accuracy 
        count = 0
        for i in xrange(self.Y.shape[0]):
            if predict_digits[i] == self.Y[i]:
                count+=1
        predict_all = 100.0*( float(count)/self.Y.shape[0] )
        
        # calculating the percent accuracy 
        # in predicting these individual digits
        acc = np.zeros(k,float)
        for i in xrange(self.Y.shape[0]):
            if predict_digits[i] == self.Y[i]:
                for j in xrange(k):
                    if self.Y[i] == digits[j]:
                        acc[j]+=1
        predict_each = 100.0*(acc/self.digit_count())
        return predict_all,predict_each    
```

Load the predicted parameters of the logistic regression model, the $\theta$ matrix for the $k=10$ classifiers (digits). The probabilistic predictions for a particular digit is $\theta^T X$, where $X$ is the feature matrix for the train or test data set.

```python
Nmax = 200    # max number of examples per digit
Nmin = 50     # min number of examples per digit
Nstep = 50    # steps of N increment
i = Nmin
thetas_ub = []
while i <= Nmax:
    theta = np.load('thetafirst'+str(i)+'.npy')
    # we store example number N, 
    # and corresponding theta matrix
    thetas_ub.append( (i,theta) )
    i+=Nstep
```

The predicted results from the logistic regression model for the test set, using unbiased training sets of various sizes.

```python
predict_ub = Predict_Unbiased(Xtest,Ytest)
```

```python
# accuracy of the logistic regression predictions
accuracy_ub = []
for i in xrange( len(thetas_ub) ):
    accuracy_ub.append(predict_ub.accuracy(thetas_ub[i][1]))
    i+=1
```

## Logistic regression using biased training sets

In this section, we create training sets which are different for every digit. 

### Biased training sets

We create biased training sets starting from an unbiased one, say with $n$ training examples per digit. We add a fixed bias of $b$ training examples corresponding to the digit $d$ being learned. If bias $b/n\gg1$, we will fail to have a classification problem with the training set predominantly containing the digit $d$. If $|b|$ is large and $b$ is negative, we will create a training set with very few instances of digit $d$. Thus, there probably is an optimum $b$ that will help the learning process. Our approach, however, is limited here. Instead of finding such an optimum $b$, we show that biased training sets does help the learning process. We particularly study two cases, one with $n=100$ and $b=100$ and another with $n=150$ and $b=100$. The total number of training examples being used for each calculation is thus $N=nk+b$. 

### Predictions from the model

```python
class Predict_Biased(Predict):
    # we define how to calculate accuracy in this case
    # in this case, there are two types of \theta matrices
    # theta_biased contains a list of 10 \theta arrays
    # corresponding to the 10 different biased training sets
    # each \theta array is in itself a 10x785 numpy array
    # theta_unbiased is a final effort in predicting digits 
    # that remained unaccounted for using the biased method
    
    def accuracy(self,theta_biased,theta_unbiased):
        # base unbiased prediction, use the best unbiased
        # \theta matrix that you have
        digit_ub = self.pred_digits(theta_unbiased)
        
        # prediction count per digit
        k = 10 # number of classifiers
        pred_count = np.zeros(k,float)
        
        # main code calculating the prediction accuracy
        for i in xrange(k):
            pred_countold = pred_count
            digit_b = self.pred_digits(theta_biased[i]) 
            for j in xrange(self.Y.shape[0]):       
                if digit_b[j] == i and self.Y[j] == i:
                    pred_count[i]+=1                
            if np.array_equal(pred_countold, pred_count):
                digit_b[j] = digit_ub[j]
                if digit_b[j] == self.Y[j]:
                    pred_count[digit_b[j]]+=1
        pred_digits = 100.0*(pred_count)/self.digit_count()
        return np.average(pred_digits), pred_digits 
```

Loading the biased theta matrices for the two different cases.

```python
theta_ub1 = np.load('thetafirst100digall200.npz')
theta_ub2 = np.load('thetafirst150digall250.npz')
thetas100_b = []
thetas150_b = []
for i in xrange(k):
    theta1 = theta_ub1['theta'+str(i)]
    theta2 = theta_ub2['theta'+str(i)]
    # we store example number N, 
    # and corresponding theta matrix
    thetas100_b.append( theta1 )
    thetas150_b.append( theta2 )
```

Predicted results for the test set using the above two trained logistic regressions.

```python
predict_b = Predict_Biased(Xtest,Ytest)
acc100_b = predict_b.accuracy(thetas100_b,thetas_ub[3][1])
acc150_b = predict_b.accuracy(thetas150_b,thetas_ub[3][1])
```

## Correlations between handwritten digits

We thus see that the model has difficulty predicting certain digits, like $8$, while on the other hand, it is particularly good at predicting certain digits, such as $1$. This should reflect on the illegibility of our handwriting pertaining to certain digits and point to any underlying patterns theirin. 


### Correlations between digits for wrong predictions

Our next effort is to determe, if a particular digit is recognized incorrectly, whether there is an underlying pattern in this. For instance, if digit $8$ is incorrectly identified, then, whether the model predicts a particular "wrong" digit more often. Also, it will be very interesting to visually check how close the particular grey scale image of the digit resembles the wrongly predicted value. This information will hopefully give us more insight into common handwriting practices and errors. 

For every digit, our code predicts a corresponding digit which may or may not be correct. For a given digit, we find out the the distribution of our prediction over all digits. If this distribution is flat, the error is more or less random. If the distrubution has features, then it will point to a more systematic error, in which a given digit is more prone to be mis-understood for another. 

### Correlations between digits for correct predictions

If our code predicts a given digit correctly, we estimate the distribution of over all digits that are of second highest probability, and try to see if there is any underlying pattern. Actually, we hope to see a similar pattern as what we saw in the case where the digits were predicted incorrectly. 

```python
# finding digit_correlations 
# for both the wrong and correct predictions

class Digit_Correlations(Predict):
        
        k = 10 # number of classifiers or digits
        
        # finding out the correlations between digits
        # i.e., say we are considering digit 8
        # then, we find out how many times
        # the code predicts other digits instead of 8
        # for instance, in case of perfect accuracy, 
        # the code will always correctly predict digit 8 
        # with zero prediction for other digits
        # we report this prediction as a percentage               
        def correlations_wrong(self,theta):            
            digit_correlation = np.zeros((k,k),int)
            preds = []
            for i in xrange(k):
                preds.append(self.pred_digits(theta[i]))
            for i in xrange(self.Y.shape[0]):
                for j in xrange(k):
                    if self.Y[i] == j: 
                        digit_correlation[j,preds[j][i]]+=1
            
            digits_errors = np.zeros((k,k-1),float)
            for i in xrange(k):
                count = 0
                for j in xrange(k):
                    if i != j:
                        digits_errors[i,count] = digit_correlation[i,j]
                        count+=1
                digits_errors[i] = digits_errors[i]\
                    /float( np.sum(digits_errors[i]) )*100.0
                
            return digits_errors
        
        # If our code predicts a given digit correctly
        # we estimate the distribution of over all digits
        # that are of second highest probability 
        # and try to see if there is any underlying pattern
        # we report this as a percentage distribution        
        def correlations_next(self,theta):
            digits_inorder = np.zeros((k,k),int)
            preds = []
            for i in xrange(k):
                digits = np.argsort( -self.probs(theta[i]) )
                preds.append(digits)
            for i in xrange(self.Y.shape[0]):
                for j in xrange(k):
                    if self.Y[i] == j and preds[j][i,0] ==  j:            
                        digits_inorder[j,preds[j][i,1]]+=1

            digits_next = np.zeros((k,k-1),float)
            for i in xrange(k):
                count = 0
                for j in xrange(k):
                    if i != j:
                        digits_next[i,count] = digits_inorder[i,j]
                        count+=1
                digits_next[i] = digits_next[i]\
                    /float( np.sum(digits_next[i]) )*100.0
            return digits_next
```

```python
correlations = Digit_Correlations(Xtest,Ytest)
corr_wrong = correlations.correlations_wrong(thetas100_b)
corr_next = correlations.correlations_next(thetas100_b)
```

Defining the pie chart features for viasualizing the digit correlations.

```python
def pie_plot(size,colors,label,explode):
    plotpie = plt.pie(size,
                      colors=colors,
                      labels=label,
                      explode=explode,
                      autopct='%1i%%',
                      shadow=True,
                      startangle=90)
    
# labels for pie plots
labels_pie = []
for i in xrange(k):
    label = []
    for j in xrange(k):
        if i != j:
            label.append(str(j))
    labels_pie.append(label)

# colors of choice for the pie plots
colors = ['yellowgreen', 'gold', 'lightskyblue', 
          'lightcoral','lightpink','peachpuff',
          'lightseagreen','mistyrose','aquamarine']
```

Plotting the digit correlations for failed predictions as pie charts.

```python
# we want to explode two portions 
# of highest percentages
explode = np.zeros((k,k-1),float)
for i in xrange(k):
    ones = np.zeros(k-1,float)
    index = np.argsort(corr_wrong[i])
    ones[index[k-2]], ones[index[k-3]]= 0.1,0.1
    explode[i] = explode[i] + ones

plt.rcParams['font.size'] = 14
plt.figure(figsize=(17, 25))


for i in xrange(k):
    plt.subplot(5,2,i+1)
    pie_plot(corr_wrong[i],colors,labels_pie[i],explode[i])
    plt.title('digit'+str(i), fontsize=16)

plt.show()
```

The plots show the distribution of failed predictions from our code pertaining to every digit. We see that when there is an error in identification, there is a strong chance that the code predicts with more probability a specific other digit. For instance, digit $4$ is more likely to be mis-predicted as digit $9$.

From these plots, we can broadly infer correlations between different handwritten digits. In fact, these correlations are not simply an artefact of the code, but reflects upon common handwriting practices and errors. We can indeed exploit these correlations to write more involved code to further resolve these digits in case of ambiguity.

Plotting the digit correlations for correct predictions as pie charts.

```python
# we want to explode two portions 
# of highest percentages
explode = np.zeros((k,k-1),float)
for i in xrange(k):
    ones = np.zeros(k-1,float)
    index = np.argsort(corr_next[i])
    ones[index[k-2]], ones[index[k-3]]= 0.1,0.1
    explode[i] = explode[i] + ones

plt.rcParams['font.size'] = 14
plt.figure(figsize=(17, 25))


for i in xrange(k):
    plt.subplot(5,2,i+1)
    pie_plot(corr_next[i],colors,labels_pie[i],explode[i])
    plt.title('digit'+str(i), fontsize=16)

plt.show()
```

The pattern in overwhelmingly clear in these plots. Thus, there are indeed strong correlations between certain digits. These correlations to a large extent lead to prediction errors. Most interestingly, these plots capture quite nicely the visual similarity between different handwritten digits.
